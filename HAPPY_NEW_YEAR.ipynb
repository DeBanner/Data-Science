```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
def make_hny(N=6000, rseed=20):
    fig, ax = plt.subplots(figsize=(7, 8))
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
    ax.axis('off')
    ax.text(0.5, 0.8, 'WELCOME TO', va= 'center', ha='center', weight='bold', size=65)
    ax.text(0.5, 0.4, 'YEAR 2024', va='center', ha='center', weight='bold', size=65)

    fig.savefig('hny.png')
    plt.close(fig)

    # Open this PNG and draw random points from it
    from matplotlib.image import imread
    data = imread('hny.png')[::-1, :, 0].T
    rng = np.random.RandomState(rseed)
    X = rng.rand(4 * N, 2)
    i, j = (X * data.shape).astype(int).T
    mask = (data[i, j] < 1)
    X = X[mask]
    X[:, 0] *= (data.shape[0] / data.shape[1])
    X = X[:N]
    return X[np.argsort(X[:, 0])]
```

```python
sns.set_style('whitegrid')
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=plt.cm.tab10.colors)
plt.style.use('classic')
plt.style.use('seaborn-deep')
plt.grid(color='red', linestyle='--', linewidth=0.5)
X = make_hny(6000)
colorize = dict(c=X[:, 0], cmap=plt.cm.get_cmap('rainbow', 5))
plt.scatter(X[:, 0], X[:, 1], **colorize)
plt.title('HAPPY NEW YEAR 2024', fontdict={'fontsize': 16, 'fontweight': 'bold', 'color': 'purple'})
plt.xlabel('YOUR YEAR OF EXPLOITS', fontdict={'fontsize': 16, 'fontweight': 'bold', 'color': 'red'})
plt.axis('equal');
```

```python
plt.style.use('classic')
plt.style.use('seaborn-deep')
def rotate(X, angle):
    theta = np.deg2rad(angle)
    R = [[np.cos(theta), np.sin(theta)],
    [-np.sin(theta), np.cos(theta)]]
    return np.dot(X, R)

X2 = rotate(X, 20) + 5
plt.scatter(X2[:, 0], X2[:, 1], **colorize)
plt.title('HAPPY NEW YEAR', fontdict={'fontsize': 16, 'fontweight': 'bold', 'color': 'green'})
plt.xlabel('IN RADIANCE', fontdict={'fontsize': 16, 'fontweight': 'bold', 'color': 'royalblue'})
plt.axis('equal');

sizes = np.random.rand(len(X2)) * 100
colors = plt.cm.get_cmap('viridis')(np.linspace(0, 1, len(X2)))

plt.scatter(X2[:, 0], X2[:, 1], s=sizes, c=colors, alpha=0.5);
```

```python
class Colors:
    RED = '\033[91m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    BLUE = '\033[94m'
    RESET = '\033[0m'

def get_line_color(index):
    colors = [Colors.RED, Colors.GREEN, Colors.YELLOW, Colors.BLUE]
    return colors[index % len(colors)]

newyr = []

for y in range(14, -14, -1):
  nxt = ""
  for x in range(-30, 30):
    f = ((x * 0.05)**2 + (y * 0.1)** 2 -1) ** 3 - (x * 0.05) ** 2 * (y * 0.1) ** 3
    nxt += " HAPPY NEW YEAR 2024"[(x - y) % len(" HAPPY NEW YEAR 2024")] if f <= 0 else " "
  newyr.append(nxt)
print(Colors.GREEN +"\n".join(newyr))
```

```python
for index, line in enumerate(newyr):
    color = get_line_color(index)
    print(color + line + Colors.RESET)
```
